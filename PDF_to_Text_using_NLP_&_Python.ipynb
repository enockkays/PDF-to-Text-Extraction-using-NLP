# **Read PDF File**

```python
# The function Open allows you to read the file.
pdfFileObj = open(filename, 'rb')

# The pdfReader variable is a readable object that will be parsed.
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

# Discerning the number of pages will allow us to parse through all the pages.
num_pages = pdfReader.numPages
count = 0
text = ""

# The while loop will read each page.
while count < num_pages:
  pageObj = pdfReader.getPage(count)
  count +=1
  text += pageObj.extractText()

# This if statement exits to check if the above libray returned words.
# It's done because PyPDF2 connot read scanned files.

if text != "":
  text = text
# If the above returns as False, we run the OCR library returned words.
# convert scanned/image based PDF files into text. 
else:
  textract.process(filename, method = 'tesseract', language = 'eng')
```

# **Tokenising, Word Stemming & Removing Punctuation**

```python
str_text = text

# The word_tokenize() function will break our text phrases into individual words.
# We also use the function lower to convert our text to lower case.
tokens = word_tokenize(str_text.lower())

# Stemming is the process of reducing a word to its word stem that affixes to suffixes 
# and prefixes or to the roots of words known as a lemma. 
stemmer = PorterStemmer()

filtered_words = []

for word in tokens:
    if word not in stop_words and word.isalpha(): # We use the function isAlpha to remove punction marks.
        stemmed_word = stemmer.stem(word)
        filtered_words.append(stemmed_word)

print(filtered_words)
print('Done!')
```
